```python
import cv2
import exifread
import glob
import pandas as pd
import re
```

```python
files = glob.glob('images/*jpeg')
```

```python
imtags = []

for file in files:
    #im = cv2.imread(file)
    
    # Open image file for reading (binary mode)
    f = open(file, 'rb')

    # Return Exif tags
    tags = exifread.process_file(f)
    y, mo, d, h, m, s = re.split(':| ', tags['Image DateTime'].__str__())
    
    imtags.append(tags)
imtags = pd.DataFrame(imtags)
```
